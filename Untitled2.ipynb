```python
# =========================================================
# EfficientNet Transfer Learning on Food11
# Full notebook code: Feature Extraction + Fine-Tuning
# Google Colab / GPU / TPU friendly
# =========================================================

# =========================
# 1. Imports
# =========================
import os
import glob
import random
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

print("TensorFlow version:", tf.__version__)
print("Available GPUs:", tf.config.list_physical_devices('GPU'))


# =========================
# 2. Detect TPU / GPU / CPU
# =========================
try:
    tpu = tf.distribute.cluster_resolver.TPUClusterResolver()
    print("Running on TPU:", tpu.master())
    tf.config.experimental_connect_to_cluster(tpu)
    tf.tpu.experimental.initialize_tpu_system(tpu)
    strategy = tf.distribute.TPUStrategy(tpu)
except:
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        print("Running on GPU")
        strategy = tf.distribute.MirroredStrategy()
    else:
        print("Running on CPU")
        strategy = tf.distribute.get_strategy()

print("Number of replicas:", strategy.num_replicas_in_sync)


# =========================
# 3. Download / Extract Dataset
# =========================
def get_data_extract():
    if "dataset" in os.listdir():
        print("Dataset already exists")
    else:
        print("Downloading the data...")
        !wget -O food-data.zip https://www.kaggle.com/api/v1/datasets/download/trolukovich/food11-image-dataset
        print("Dataset downloaded!")
        print("Extracting data...")
        !mkdir dataset
        !unzip -q food-data.zip -d dataset
        print("Extraction done!")

get_data_extract()


# =========================
# 4. Collect Image Paths
# =========================
all_paths = glob.glob("dataset/*/*/*.jpg")
print("Total images found:", len(all_paths))

# Optional: shuffle once before any processing
random.seed(42)
random.shuffle(all_paths)

# Extract labels from folder names
all_labels = [p.split("/")[-2] for p in all_paths]

class_names = sorted(list(set(all_labels)))
num_classes = len(class_names)

print("Classes:", class_names)
print("Number of classes:", num_classes)


# =========================
# 5. Train / Validation Split
# =========================
train_paths, val_paths, train_labels, val_labels = train_test_split(
    all_paths,
    all_labels,
    test_size=0.2,
    random_state=42,
    stratify=all_labels
)

print("Train samples:", len(train_paths))
print("Validation samples:", len(val_paths))


# =========================
# 6. Encode Labels
# =========================
label_to_index = {label: idx for idx, label in enumerate(class_names)}
index_to_label = {idx: label for label, idx in label_to_index.items()}

train_labels_idx = np.array([label_to_index[label] for label in train_labels], dtype=np.int32)
val_labels_idx   = np.array([label_to_index[label] for label in val_labels], dtype=np.int32)


# =========================
# 7. Config
# =========================
IMG_SIZE = 224
BATCH_SIZE = 32 * strategy.num_replicas_in_sync
AUTOTUNE = tf.data.AUTOTUNE
SEED = 42

print("IMG_SIZE:", IMG_SIZE)
print("BATCH_SIZE:", BATCH_SIZE)


# =========================
# 8. Preprocessing & Augmentation
# =========================
from tensorflow.keras.applications.efficientnet import preprocess_input

data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
], name="data_augmentation")


# =========================
# 9. Data Loading Functions
# =========================
def load_image(path, label):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = preprocess_input(image)
    return image, label

def make_dataset(paths, labels, training=True):
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))

    if training:
        ds = ds.shuffle(buffer_size=len(paths), seed=SEED)

    ds = ds.map(load_image, num_parallel_calls=AUTOTUNE)

    if training:
        ds = ds.map(
            lambda x, y: (data_augmentation(x, training=True), y),
            num_parallel_calls=AUTOTUNE
        )

    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(AUTOTUNE)
    return ds

train_ds = make_dataset(train_paths, train_labels_idx, training=True)
val_ds   = make_dataset(val_paths, val_labels_idx, training=False)


# =========================
# 10. Visualize Sample Batch
# =========================
images, labels = next(iter(train_ds))
print("Image batch shape:", images.shape)
print("Label batch shape:", labels.shape)

plt.figure(figsize=(12, 8))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    img = images[i].numpy()
    img_disp = (img - img.min()) / (img.max() - img.min() + 1e-8)
    plt.imshow(img_disp)
    plt.title(index_to_label[int(labels[i].numpy())])
    plt.axis("off")
plt.tight_layout()
plt.show()


# =========================
# 11. Plot Function
# =========================
def plot_history(history, title="Training History"):
    acc = history.history["accuracy"]
    val_acc = history.history["val_accuracy"]
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]
    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, label="Train Accuracy")
    plt.plot(epochs, val_acc, label="Validation Accuracy")
    plt.title(f"{title} - Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, label="Train Loss")
    plt.plot(epochs, val_loss, label="Validation Loss")
    plt.title(f"{title} - Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.tight_layout()
    plt.show()


# =========================
# 12. Build Feature Extraction Model
# =========================
def build_feature_extractor(num_classes):
    base_model = keras.applications.EfficientNetB0(
        include_top=False,
        weights="imagenet",
        input_shape=(IMG_SIZE, IMG_SIZE, 3)
    )
    base_model.trainable = False

    inputs = keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = base_model(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = keras.Model(inputs, outputs, name="EfficientNetB0_FeatureExtraction")
    return model, base_model


# =========================
# 13. Callbacks for Feature Extraction
# =========================
feature_callbacks = [
    keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=4,
        restore_best_weights=True
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.2,
        patience=2,
        min_lr=1e-7
    ),
    keras.callbacks.ModelCheckpoint(
        "best_feature_extraction.keras",
        monitor="val_accuracy",
        save_best_only=True,
        verbose=1
    )
]


# =========================
# 14. Compile Feature Extraction Model
# =========================
with strategy.scope():
    feature_model, base_model = build_feature_extractor(num_classes)
    feature_model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

feature_model.summary()


# =========================
# 15. Train Feature Extraction
# =========================
history_feature = feature_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=feature_callbacks
)


# =========================
# 16. Evaluate Feature Extraction
# =========================
feature_eval = feature_model.evaluate(val_ds, verbose=1)
print("Feature Extraction - Validation Loss:", feature_eval[0])
print("Feature Extraction - Validation Accuracy:", feature_eval[1])

plot_history(history_feature, "Feature Extraction")


# =========================
# 17. Fine-Tuning Setup
# Unfreeze last N layers only
# =========================
UNFREEZE_LAST_N = 30

base_model.trainable = True

for layer in base_model.layers[:-UNFREEZE_LAST_N]:
    layer.trainable = False

for layer in base_model.layers[-UNFREEZE_LAST_N:]:
    layer.trainable = True

print(f"Total base model layers: {len(base_model.layers)}")
print(f"Unfrozen last {UNFREEZE_LAST_N} layers")


# =========================
# 18. Fine-Tuning Callbacks
# =========================
fine_tune_callbacks = [
    keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=4,
        restore_best_weights=True
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.2,
        patience=2,
        min_lr=1e-7
    ),
    keras.callbacks.ModelCheckpoint(
        "best_fine_tuned.keras",
        monitor="val_accuracy",
        save_best_only=True,
        verbose=1
    )
]


# =========================
# 19. Recompile for Fine-Tuning
# Important: use smaller LR
# =========================
with strategy.scope():
    feature_model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-5),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

feature_model.summary()


# =========================
# 20. Train Fine-Tuning
# =========================
history_finetune = feature_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=fine_tune_callbacks
)


# =========================
# 21. Evaluate Fine-Tuning
# =========================
finetune_eval = feature_model.evaluate(val_ds, verbose=1)
print("Fine-Tuning - Validation Loss:", finetune_eval[0])
print("Fine-Tuning - Validation Accuracy:", finetune_eval[1])

plot_history(history_finetune, "Fine-Tuning")


# =========================
# 22. Final Comparison
# =========================
print("===== Final Comparison =====")
print(f"Feature Extraction Accuracy : {feature_eval[1]:.4f}")
print(f"Fine-Tuning Accuracy        : {finetune_eval[1]:.4f}")
print(f"Feature Extraction Loss     : {feature_eval[0]:.4f}")
print(f"Fine-Tuning Loss            : {finetune_eval[0]:.4f}")


# =========================
# 23. Predictions on Validation Set
# =========================
pred_probs = feature_model.predict(val_ds)
pred_classes = np.argmax(pred_probs, axis=1)

true_classes = np.concatenate([y.numpy() for _, y in val_ds])

print("Prediction shape:", pred_classes.shape)
print("True labels shape:", true_classes.shape)


# =========================
# 24. Classification Report
# =========================
print(classification_report(true_classes, pred_classes, target_names=class_names))


# =========================
# 25. Confusion Matrix (matplotlib only)
# =========================
cm = confusion_matrix(true_classes, pred_classes)

plt.figure(figsize=(12, 10))
plt.imshow(cm, interpolation='nearest')
plt.title("Confusion Matrix")
plt.colorbar()

tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=90)
plt.yticks(tick_marks, class_names)

threshold = cm.max() / 2.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(
            j, i, format(cm[i, j], 'd'),
            horizontalalignment="center",
            color="white" if cm[i, j] > threshold else "black"
        )

plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.tight_layout()
plt.show()


# =========================
# 26. Show Some Predictions
# =========================
val_images, val_batch_labels = next(iter(val_ds))
sample_preds = feature_model.predict(val_images[:9], verbose=0)
sample_pred_labels = np.argmax(sample_preds, axis=1)

plt.figure(figsize=(12, 8))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    img = val_images[i].numpy()
    img_disp = (img - img.min()) / (img.max() - img.min() + 1e-8)
    plt.imshow(img_disp)

    true_label = index_to_label[int(val_batch_labels[i].numpy())]
    pred_label = index_to_label[int(sample_pred_labels[i])]

    plt.title(f"T: {true_label}\nP: {pred_label}")
    plt.axis("off")

plt.tight_layout()
plt.show()


# =========================
# 27. Save Final Model
# =========================
feature_model.save("efficientnet_food11_final.keras")
print("Final model saved as efficientnet_food11_final.keras")


# =========================
# 28. Save Training Curves as Images
# =========================
def save_history_plot(history, title, filename):
    acc = history.history["accuracy"]
    val_acc = history.history["val_accuracy"]
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]
    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, label="Train Accuracy")
    plt.plot(epochs, val_acc, label="Validation Accuracy")
    plt.title(f"{title} - Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, label="Train Loss")
    plt.plot(epochs, val_loss, label="Validation Loss")
    plt.title(f"{title} - Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches="tight")
    plt.show()

save_history_plot(history_feature, "Feature Extraction", "feature_extraction_metrics.png")
save_history_plot(history_finetune, "Fine-Tuning", "fine_tuning_metrics.png")


# =========================
# 29. Optional: MLflow Tracking
# Uncomment if needed
# =========================
"""
!pip install mlflow dagshub -q

import mlflow
import mlflow.tensorflow

mlflow.set_experiment("EfficientNet_Food11")

with mlflow.start_run(run_name="feature_extraction_b0"):
    mlflow.log_param("model", "EfficientNetB0")
    mlflow.log_param("img_size", IMG_SIZE)
    mlflow.log_param("batch_size", BATCH_SIZE)
    mlflow.log_param("learning_rate", 1e-3)
    mlflow.log_param("frozen_base", True)
    mlflow.log_metric("val_accuracy", float(feature_eval[1]))
    mlflow.log_metric("val_loss", float(feature_eval[0]))
    mlflow.log_artifact("feature_extraction_metrics.png")
    mlflow.tensorflow.log_model(feature_model, "feature_extraction_model")

with mlflow.start_run(run_name="fine_tuning_b0"):
    mlflow.log_param("model", "EfficientNetB0")
    mlflow.log_param("img_size", IMG_SIZE)
    mlflow.log_param("batch_size", BATCH_SIZE)
    mlflow.log_param("learning_rate", 1e-5)
    mlflow.log_param("unfreeze_last_n", UNFREEZE_LAST_N)
    mlflow.log_metric("val_accuracy", float(finetune_eval[1]))
    mlflow.log_metric("val_loss", float(finetune_eval[0]))
    mlflow.log_artifact("fine_tuning_metrics.png")
    mlflow.tensorflow.log_model(feature_model, "fine_tuned_model")
"""


# =========================
# 30. Simple Printed Analysis
# =========================
print("\n===== ANALYSIS =====")
print("1. Feature Extraction:")
print("- Base EfficientNetB0 was frozen.")
print("- Only the classification head was trained.")
print("- This is faster and reduces overfitting risk.")
print("- Usually converges quickly but may plateau earlier.")

print("\n2. Fine-Tuning:")
print("- Last layers of EfficientNetB0 were unfrozen.")
print("- Model was retrained with a smaller learning rate.")
print("- This helps adapt pretrained features to Food11.")
print("- Usually gives better validation accuracy, but overfitting risk is higher.")

print("\n3. Generalization / Convergence / Overfitting:")
print("- Feature extraction usually generalizes well with limited training cost.")
print("- Fine-tuning often improves task-specific performance.")
print("- If validation loss rises while training accuracy keeps rising, overfitting is happening.")
print("- EarlyStopping and ReduceLROnPlateau help stabilize training.")
```
